## 1. K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
;# Génération de données
X, y,centers = make_blobs(n_samples=200,  cluster_std=0.4,random_state=0, return_centers=True)

centers

```

```python
KMeans
```

```python
model = KMeans(n_clusters=3)
model.fit(X)
model.predict(X) # donnner de test
plt.scatter(X[:,0], X[:,1], c=model.predict(X))
plt.colorbar()

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c='r')


```

```python
model.cluster_centers_
```

### Elbow Method

```python
print(model.inertia_)
```

```python
inertia = []
K_range = range(1, 20)
for k in K_range:
    model = KMeans(n_clusters=k).fit(X)
    inertia.append(model.inertia_)

plt.plot(K_range, inertia)
plt.xlabel('nombre de clusters')
plt.ylabel('Cout du modele (Inertia)')
```

## 2. Detection d'anomalies avec Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
X, y = make_blobs(n_samples=50, centers=1, cluster_std=0.1, random_state=0)
X[-1,:] = np.array([2.25, 5])

plt.scatter(X[:,0], X[:, 1])
```

```python
model = IsolationForest(n_estimators=1000,contamination=0.06, random_state=0)
model.fit(X)
plt.scatter(X[:,0], X[:, 1], c=model.predict(X))
plt.colorbar()
```

```python
model.predict(X)
```

### Application : Digits Outliers

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits.data.shape
images = digits.images
X = digits.data
y = digits.target
#print(X.shape)
plt.imshow(images[1200])

plt.title(y[1200])

```

```python
model = IsolationForest(random_state=0, contamination=0.02)
model.fit(X)
c=model.predict(X)
outliers = model.predict(X) == -1
plt.figure(figsize=(12, 3))
for i in range(36):
    plt.subplot(1, 36, i+1)
    plt.imshow(images[outliers][i])
    plt.title(y[outliers][i])
    #plt.title([outliers][i])
```

## 3. Reduction de dimension avec PCA

### Visualisation 2D

```python
from sklearn.decomposition import PCA

model = PCA(n_components=2)
x_pca=model.fit_transform(X)
print(x_pca.shape)
print(X.shape)
```

```python
plt.scatter(x_pca[:,0], x_pca[:,1], c=y)
plt.colorbar()
print(y.shape)
```

```python
model.components_.shape
```

```python
plt.figure()
plt.xlim(-30, 30)
plt.ylim(-30, 30)

for i in range(1000):
    plt.text(x_pca[i,0], x_pca[i,1], str(y[i]))
```

### Compression de données

```python
n_dims = X.shape
n_dims
```

```python
n_dims = X.shape[1]
model = PCA(n_components=n_dims)
X_compress = model.fit_transform(X)
variances = model.explained_variance_ratio_
plt.bar(range(n_dims), np.cumsum(variances))

meilleur_dims = np.argmax(np.cumsum(variances) >= 0.99)



plt.hlines(0.99, 0, meilleur_dims, colors='r')
plt.vlines(meilleur_dims, 0, 0.99, colors='r')
n_dims
np.argmax(np.cumsum(variances)>=0.99)
```

```python
model = PCA(n_components=0.1)
X_compress = model.fit_transform(X)
print(model.n_components_)
X_decompress = model.inverse_transform(X_compress)

plt.subplot(1, 2, 1)
plt.imshow(X[0,:].reshape((8,8)), cmap='gray')
plt.title('Initial')
plt.subplot(1, 2, 2)
plt.imshow(X_decompress[0,:].reshape((8,8)), cmap='gray')
plt.title('Compressé')
```
